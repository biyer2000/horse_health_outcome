```python
df_train = df_train.drop(columns=['id', 'hospital_number'])
df_test = df_test.drop(columns=['id', 'hospital_number'])
```

```python
df_train
```

```python
from sklearn.utils import resample

lived_class = df_train[df_train['outcome'] == 'lived']
died_class = df_train[df_train['outcome'] == 'died']
euthanized_class = df_train[df_train['outcome'] == 'euthanized']

minority_died_class = resample(died_class, replace=True, n_samples=len(lived_class), random_state=42)
minority_euthanized_class = resample(euthanized_class, replace=True, n_samples=len(lived_class), random_state=42)

df_train = pd.concat([lived_class, minority_died_class, minority_euthanized_class])

class_distribution = df_train['outcome'].value_counts()
print("Class distribution after upsampling:", class_distribution)
```

```python
from sklearn.preprocessing import LabelEncoder

def label_encoder(data):
  #changing categorical variables to numerical

    obj_cols = data.select_dtypes(include=['object']).columns

    label_encoder = LabelEncoder()

    for col in obj_cols:
        data[col + '_encoded'] = label_encoder.fit_transform(data[col])

    return data
```

```python
df_train = label_encoder(df_train)
df_train
```

```python
df_test = label_encoder(df_test)
df_test
```

```python
def drop_object_columns(data):


    object_columns = data.select_dtypes(include=['object'])
    data.drop(columns=object_columns, inplace=True)
    return data
```

```python
drop_object_columns(df_train)
df_train.head(5)
```

```python
drop_object_columns(df_test)
df_test.head(5)
```

```python
from sklearn.preprocessing import StandardScaler

def standard_scale(data):
  #scaling the values with mean of 0 and sd of 1


    scaler = StandardScaler()

    data_z = scaler.fit_transform(data)
    return data_z
```

```python
df_train_z = standard_scale(df_train)
df_train_z = pd.DataFrame(df_train_z, columns=df_train.columns)
df_train_z
```

```python
df_test_z = standard_scale(df_test)
df_test_z = pd.DataFrame(df_test_z, columns=df_test.columns)
df_test_z
```

```python
cor = df_train_z.corr()
cor_target = abs(cor["outcome_encoded"])

relevant_features = cor_target[cor_target>0.5]
relevant_features
```

```python
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
models = [RandomForestClassifier(),GradientBoostingClassifier(),
          DecisionTreeClassifier(),KNeighborsClassifier(),
         GaussianNB()]

```

```python
from sklearn.model_selection import train_test_split

x = df_train.drop(columns=["outcome_encoded"])
y = df_train["outcome_encoded"]

# split train set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

name = []
accuracy = []

for model in models:
  #trying different models
    name.append(type(model).__name__)
    model.fit(x_train,y_train)
    prediction = model.predict(x_test)
    accuracyscore = accuracy_score(prediction, y_test)
    accuracy.append(accuracyscore)

dic = {'Name':name,'Accuracy':accuracy}
model_df = pd.DataFrame(dic)
model_df
```

```python
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

rf_classifier.fit(x_train, y_train)

y_pred = rf_classifier.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy:", accuracy)
```
